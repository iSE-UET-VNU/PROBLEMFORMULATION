```python
import os
import zipfile
import glob
from tqdm import tqdm
import re
paths = []
for dirpath, dirname, filenames in os.walk('./csv_test'):
  for filename in filenames:
    path = os.path.join(dirpath, filename)
    with zipfile.ZipFile(path, 'r') as zip_file:
      name_list = zip_file.namelist()
      paths.append(path)
```

```python
import pandas as pd
file_df = pd.read_csv('/content/csv_test.csv')

selected_paths = []
for file in paths:
  slug = file.split('/')[-1]
  if slug in file_df.loc[(file_df['train_file'].str.contains('.csv') & ~file_df['train_file'].str.contains(','))]['file_name'].to_list():
    selected_paths.append(file)
```

```python
selected_paths_df = pd.DataFrame(selected_paths, columns = ['zip_path'])
selected_paths_df['folder_path'] = selected_paths_df['zip_path'].apply(lambda x: os.path.join('./csv_test_extracted', x.split('/')[-1]).replace('.zip', ''))
selected_paths_df['zip_name'] = selected_paths_df['zip_path'].apply(lambda x: x.split('/')[-1])

selected_paths_df = pd.merge(selected_paths_df, file_df, left_on = 'zip_name', right_on = 'file_name')
selected_paths_df.drop(['zip_name', 'file_name', 'flag'], axis = 1, inplace = True)
```

```python
import subprocess
from concurrent.futures import ThreadPoolExecutor, as_completed

def extract_zip_file(path):
  output_path = os.path.join('./csv_test_extracted', path.split('/')[-1].split('.')[0])
  subprocess.run(['7z', 'x', '-mmt=on', path, '-o' + output_path], stdout = subprocess.DEVNULL, stderr = subprocess.DEVNULL)


with ThreadPoolExecutor(max_workers = 10) as executor:
  futures = [executor.submit(extract_zip_file, path) for path in selected_paths_df['zip_path']]
  for _ in tqdm(as_completed(futures), total = len(futures), desc = 'extracting files', unit = 'file'):
    pass
```

```python
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, SystemMessage
from langchain_openai import ChatOpenAI
from langchain_core.output_parsers import JsonOutputParser
from typing import List, Optional, TypedDict, Literal
import json
import pandas as pd


MAX_TURN = 20

class TargetOpportunity(TypedDict):
    target_name: str
    target_description: str
    business_value: str
    ml_problem_type: str
    feature_engineering_hints: List[str]
    confidence_score: float

class Insight(TypedDict):
    insight_type: str
    insight: str
    reason: str
    target_opportunities: List[TargetOpportunity]

class DiscoveryState(TypedDict):
    dataset_schema: str
    profiling_summary: str
    explored_perspectives: List[str]
    all_insights: List[Insight]
    discovered_targets: List[TargetOpportunity]
    next_plan: Optional[str]
    stop_reason: Optional[str]
    turn: int


# 2. INITIALIZE LLM & JSON PARSER
# ===========================================
llm = ChatOpenAI(model="gpt-4o-mini", temperature=0.1)
json_parser = JsonOutputParser()

# TARGET-FOCUSED PLANNER NODE
# ===========================================
def planner_node(state: DiscoveryState) -> DiscoveryState:
    dataset_schema = state["dataset_schema"]
    profiling_summary = state["profiling_summary"]
    explored = state["explored_perspectives"]
    discovered_targets = state.get("discovered_targets", [])

    if not explored:
        prompt = f"""
You are an ML Strategy Expert specialized in Target Variable Discovery.

MISSION CRITICAL: Discover insights to generate NEW TARGET COLUMNS of high value for Machine Learning.

TARGET-DISCOVERY METHODOLOGY (choose the next step):

**1. BUSINESS OUTCOME ANALYSIS**
   - Plan: "Analyze potential business outcomes from the current data"
   - Target Goal: Identify KPIs, metrics, or outcomes that can be predicted/classified
   - Examples: Customer satisfaction score, failure probability, performance rating

**2. FEATURE COMBINATION OPPORTUNITIES**
   - Plan: "Explore combinations that create meaningful targets"
   - Target Goal: Ratios, indices, composite scores
   - Examples: Efficiency_score = Output/Input, Risk_level = f(multiple factors)

**3. BEHAVIORAL PATTERN TARGETS**
   - Plan: "Identify behavioral patterns to predict"
   - Target Goal: User actions, preferences, tendencies
   - Examples: Will_churn, Preferred_category, Usage_frequency_level

**4. THRESHOLD-BASED CLASSIFICATION TARGETS**
   - Plan: "Identify natural thresholds for classification"
   - Target Goal: Binary/multi-class labels derived from continuous variables
   - Examples: High_performer (top 20%), Outlier_behavior, Risk_category

**5. TEMPORAL PATTERN TARGETS**
   - Plan: "Explore time-based outcomes to predict"
   - Target Goal: Future states, trends, cycles
   - Examples: Next_month_performance, Seasonal_demand_type

**6. DOMAIN-SPECIFIC VALUE TARGETS**
   - Plan: "Apply domain knowledge to create specialized targets"
   - Target Goal: Industry-specific outcomes
   - Examples: Credit_worthiness, Equipment_health_score, Customer_lifetime_segment

**7. OPTIMIZATION TARGETS**
   - Plan: "Identify targets for optimization problems"
   - Target Goal: Cost minimization, efficiency maximization
   - Examples: Optimal_resource_allocation, Best_maintenance_schedule

Context:
Dataset: {dataset_schema}
Profiling: {profiling_summary}

→ Output JSON:
{{
  "plan": {{
    "next_plan": "Specific target discovery step + detailed action plan",
    "reason": "Why this step will yield valuable target opportunities"
  }}
}}
"""
    else:
        explored_steps = len(explored)
        target_count = len(discovered_targets)

        prompt = f"""
ML Strategy Expert - Target Discovery Progress Review

CURRENT STATUS:
- Completed Steps: {explored_steps}
- Target Opportunities Found: {target_count}
- Explored: {explored}

TARGET QUALITY METRICS:

- Need at least 3-5 HIGH-VALUE targets
- Each target must solve real business problems
- Must be predictable from existing features
- Must explore AT LEAST 6-8 different methodology steps
- Currently only {explored_steps} steps completed → need to continue

NEXT STEP OPTIONS:

**DISCOVERY STEPS:** (if more targets are needed)
1. Business Outcome Analysis
2. Feature Combination Opportunities
3. Behavioral Pattern Targets
4. Threshold-based Classification
5. Temporal Pattern Targets
6. Domain-specific Value Targets
7. Optimization Targets

**REFINEMENT STEPS:** (if targets exist but need improvement)
8. Target Validation & Scoring
9. Feature Engineering Deep Dive
10. Cross-target Relationship Analysis

**COMPLETION:** Use "STOP" if 10+ high-quality targets have been found

Context:
Dataset: {dataset_schema}
Current targets: {discovered_targets}

Strategy Decision - Continue discovery or refine existing targets?

→ Output JSON:
{{
  "plan": {{
    "next_plan": "Next strategic step",
    "reason": "Strategic reasoning"
  }} | "STOP"
}}
"""

    messages = [
        SystemMessage(content="You are a TARGET DISCOVERY STRATEGIST - every step must lead to actionable ML targets."),
        HumanMessage(content=prompt)
    ]

    chain = llm | json_parser
    try:
        decision = chain.invoke(messages)
        plan_obj = decision.get("plan", "STOP")
        if plan_obj == "STOP":
            next_p = "STOP"
        else:
            next_p = plan_obj.get("next_plan", "STOP")
            print(f'{next_p}')
            print("=======================FFFFFFFFFFFF=================")
            state['explored_perspectives'].append(next_p)
    except Exception as e:
        next_p = "STOP"

    return {
        **state,
        "turn": state['turn'] + 1,
        "next_plan": next_p,
        "stop_reason": None if next_p != "STOP" else "Planning completed"
    }


def worker_node(state: DiscoveryState) -> DiscoveryState:
    perspective = state["next_plan"]
    print(f"=============TURN {state['turn']}============")
    print(f"Target Discovery Focus: {perspective}")

    if not perspective or perspective == "STOP":
        return state

    dataset_schema = state["dataset_schema"]
    profiling_summary = state["profiling_summary"]
    existing_targets = state.get("discovered_targets", [])

    prompt = f"""
You are a TARGET VARIABLE ENGINEER – an expert in creating high-value target columns for ML.

CRITICAL MISSION: Execute "{perspective}" to DISCOVER TARGET OPPORTUNITIES

**TARGET DISCOVERY REQUIREMENTS:**
- Each insight MUST propose at least 1-2 SPECIFIC new target columns
- Target must NOT already exist in the current dataset
- Target must have clear business value
- Target must be predictable from existing features

**DATA CONTEXT:**
Dataset Columns: {dataset_schema}
Data Analysis: {profiling_summary}
Existing Target Ideas: {len(existing_targets)} targets discovered

**OUTPUT FORMAT - MANDATORY:**
[
  {{
    "insight_type": "Target_Discovery",
    "insight": "Detailed finding that reveals target opportunity",
    "reason": "Business justification + technical feasibility",
    "target_opportunities": [
      {{
        "target_name": "Specific_Target_Column_Name",
        "target_description": "What this target represents and measures",
        "business_value": "Why this target is valuable for business decisions",
        "ml_problem_type": "classification| regression| ranking | clustering | etc...",
        "feature_engineering_hints": ["hint1", "hint2", "hint3"],
        "confidence_score": 0.85
      }}
    ]
  }}
]

EXAMPLES OF GOOD TARGETS:
- Customer_Satisfaction_Score (from behavior patterns)
- Equipment_Failure_Risk (from sensor data combinations)
- Optimal_Price_Point (from demand patterns)
- Performance_Category (from multiple metrics)

**EXECUTE THE PLAN:** {perspective}
"""

    messages = [
        SystemMessage(content="You are a TARGET OPPORTUNITY DISCOVERER - every insight must yield concrete target variables for ML."),
        HumanMessage(content=prompt)
    ]

    chain = llm | json_parser
    try:
        insights = chain.invoke(messages)
        if not isinstance(insights, list):
            insights = []

        print(f"Generated {len(insights)} target-focused insights")
        for insight in insights:
            print(f"- {insight.get('insight', 'N/A')}")
            targets = insight.get('target_opportunities', [])
            for target in targets:
                print(f"  → TARGET: {target.get('target_name', 'N/A')}")

    except Exception as e:
        print(f"[Worker Error] {e}")
        insights = []

    new_targets = []
    for insight in insights:
        targets = insight.get('target_opportunities', [])
        new_targets.extend(targets)

    return {
        **state,
        "turn": state['turn'] + 1,
        "all_insights": state["all_insights"] + insights,
        "discovered_targets": state.get("discovered_targets", []) + new_targets,
        "explored_perspectives": state["explored_perspectives"]
    }


def should_continue(state: DiscoveryState) -> Literal["worker", "end"]:
    explored_count = len(state.get("explored_perspectives", []))
    target_count = len(state.get("discovered_targets", []))

    # Stop if:
    # 1. Maximum turns reached
    # 2. Planner decides to STOP
    # 3. Enough distinct steps explored (at least 6 different types)

    if state['turn'] >= MAX_TURN:
        return "end"
    elif state["next_plan"] == "STOP":
        return "end"
    elif explored_count >= 12:
        return "end"
    else:
        return "worker"

# ===========================================
# TARGET SUMMARY FUNCTION
# ===========================================
def display_target_opportunities(discovered_targets):
    print("\n" + "="*60)
    print("DISCOVERED TARGET OPPORTUNITIES")
    print("="*60)

    if not discovered_targets:
        print("No targets discovered yet.")
        return


    target_groups = {}
    for target in discovered_targets:
        ml_type = target.get('ml_problem_type', 'unknown')
        if ml_type not in target_groups:
            target_groups[ml_type] = []
        target_groups[ml_type].append(target)

    for ml_type, targets in target_groups.items():
        print(f"\n{ml_type.upper()} TARGETS ({len(targets)} targets)")
        print("-" * 40)

        for i, target in enumerate(targets, 1):
            print(f"{i}. {target['target_name']}")
            print(f"   {target['target_description']}")
            print(f"   Value: {target['business_value']}")
            print(f"   Hints: {', '.join(target['feature_engineering_hints'][:2])}...")
            print(f"   Confidence: {target['confidence_score']}")
            print()

    print(f"📈 TOTAL DISCOVERED: {len(discovered_targets)} target opportunities")
    print("="*60)

# -- GRAPH ---
builder = StateGraph(DiscoveryState)
builder.add_node("planner", planner_node)
builder.add_node("worker", worker_node)
builder.set_entry_point("planner")
builder.add_edge("worker", "planner")
builder.add_conditional_edges(
    "planner",
    should_continue,
    {
        "worker": "worker",
        "end": END
    }
)

graph = builder.compile()


def run_target_discovery(df, data_profiling_function):
    """
    Run target discovery system on dataset
    """
    analysis = data_profiling_function(df)

    initial_state = DiscoveryState(
        dataset_schema=str(df.columns.tolist()),
        profiling_summary=str(analysis),
        explored_perspectives=[],
        all_insights=[],
        discovered_targets=[],
        next_plan=None,
        stop_reason=None,
        turn=1
    )


    result = graph.invoke(initial_state)


    print("\n" + "="*60)
    print("TARGET DISCOVERY INSIGHTS")
    print("="*60)

    for i, ins in enumerate(result["all_insights"], 1):
        print(f"\n{i}. {ins['insight']}")
        print(f"   → Reasoning: {ins['reason']}")


    display_target_opportunities(result.get("discovered_targets", []))

    if result.get("stop_reason"):
        print(f"\nProcess completed: {result['stop_reason']}")

    return result


df = pd.read_csv('train.csv').drop('rating', axis = 1)
result = run_target_discovery(df, data_profiling_json)
```

```python
import os
import pandas as pd

TARGET_CSV = "discovered_targets_manual.csv"  

def save_current_targets_to_csv(targets: list, dataset_info: dict = None, isTrue = False):
    """
    Ghi danh sách target opportunities vào CSV (append mode).
    
    """
    if not targets:
        print(" Không có target nào để lưu.")
        return

    rows = []
    for t in targets:
        row = {
            "target_name": t.get("target_name", ""),
            "target_description": t.get("target_description", ""),
            "business_value": t.get("business_value", ""),
            "ml_problem_type": t.get("ml_problem_type", ""),
            "feature_engineering_hints": ", ".join(t.get("feature_engineering_hints", [])),
            "confidence_score": t.get("confidence_score", 0.0),
            "is_true": isTrue
        }
        if dataset_info:
            row.update({
                "folder_path": dataset_info.get("folder_path", ""),
                "train_file": dataset_info.get("train_file", ""),
                "run_time": pd.Timestamp.now().strftime("%Y-%m-%d %H:%M:%S")
            })
        rows.append(row)

    df_out = pd.DataFrame(rows)
    first_write = not os.path.exists(TARGET_CSV)

    df_out.to_csv(
        TARGET_CSV,
        index=False,
        mode='a',
        header=first_write,
        encoding='utf-8-sig'
    )

    print(f" Đã ghi {len(df_out)} dòng vào {TARGET_CSV}")
```

```python
def process_and_save(file_path: str, target: str):
   
    print(f" Đang xử lý: {file_path}")

    df = pd.read_csv(file_path)
    if  target in df.columns:
        df = df.drop(target, axis=1)
        print(" Đã bỏ cột 'Target'.")

    print(df.columns)

    result = run_target_discovery(df, data_profiling_json)
    targets = result.get("discovered_targets", [])

    if not targets:
        print(" Không phát hiện target nào.")
        return

    print(f"\n Phát hiện {len(targets)} target opportunities.")

    
    user_input = input("\n Bạn có muốn ghi kết quả vào CSV không? (y/n): ").strip().lower()
    isTru = False
    if user_input == 'y':
      isTru = True


    save_current_targets_to_csv(
            targets,
            dataset_info={
                "folder_path": os.path.dirname(file_path),
                "train_file": os.path.basename(file_path),

            },
            isTrue= isTru
        )
    return result 
```
